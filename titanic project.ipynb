# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading Dataset
titanic = pd.read_csv('titanic.csv')
```

```python
titanic
```

# Removing Name, Ticket, Cabin Columns

```python
titanic = titanic.drop(['Name', 'Ticket', 'Cabin'], axis=1)
```

```python
titanic
```

# Data Wrangling

```python
# Null Values
titanic.isnull().sum()
```

```python
#Fill Null values in Age
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
```

```python
titanic.isnull().sum()
```

```python
titanic
```

```python
titanic.isnull().sum()
```

```python
#Fill Null values in Embarked using mode
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)
```

```python
titanic
```

```python
# No Null Values
titanic.isnull().sum()
```

```python
#No Duplicates
titanic.duplicated()
```

```python
titanic['Sex'].unique()
```

```python
#Replacing male with 0 and female with in Gender
titanic['Sex'].replace({'male': 0, 'female': 1}, inplace=True)
```

```python
titanic
```

```python
# Create new column of comparison making comparison of age and fare
titanic['Comparison'] = ['Young and Rich' if (age < 30 and fare > 10) else 'Older and Poor' for age, fare in zip(titanic['Age'], titanic['Fare'])]
```

```python
titanic
```

```python
x = titanic['Age']
y = titanic['Fare']
```

```python
plt.scatter(x, y, edgecolor='green' ,color='red')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
colors= ['green' if age > 40 else 'red' for age in x]
plt.scatter(x,y, color= colors)

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.show()
```

```python
color = [
    'red' if age > 65 else 
    'green' if age > 55 else 
    'blue' if age > 45 else 
    'black' if age > 35 else 
    'brown' if age > 25 else 
    'pink' if age > 15 else 
    'yellow' if age > 5 else 
    'purple'  
    for age in x
]

# Scatter plot with color argument corrected
plt.scatter(x, y, c=color)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.show()
```

```python
plt.bar(x,y, color=['red', 'green', 'blue'])
plt.show()
```

```python
class_counts = titanic['Pclass'].value_counts()

class_counts.plot(kind='bar', color=['blue', 'green', 'red'])

plt.xlabel('Passenger class')
plt.ylabel('Number of passengers')
plt.title('Number of passenger by class')

plt.show()
```

```python
# Get unique classes from the dataset, ensuring proper order
unique_classes = sorted(titanic['Pclass'].unique())

# Get the counts for each class in the same order
class_counts = [titanic[titanic['Pclass'] == cls].shape[0] for cls in unique_classes]

# Define labels and colors dynamically
labels = [f"Class {cls}" for cls in unique_classes]
colors = ['blue', 'green', 'red']

# Create the pie chart
plt.pie(class_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Add a title
plt.title('Number of Passengers by Class')

# Display the chart
plt.show()
```

```python
class_counts = titanic['Survived'].value_counts()

class_counts.plot(kind='bar', color=['red', 'green'])


plt.ylabel('Number of passengers')
plt.title('Survived vs Non Survived')
plt.xticks(ticks=(0, 1), labels=['Non-Survived', 'Survived'])
plt.show()
```
