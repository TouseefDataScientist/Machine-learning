```python
import matplotlib.pyplot as plt
```

```python
titanic =pd.read_csv("titanic.csv")
```

```python
x=titanic['Age']
y=titanic['Fare']
```

```python
plt.scatter(x,y, edgecolors='red', facecolor='green')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
```

```python

x=titanic['Age']
y=titanic['Fare']
color = ['red' if age > 30 else 'green' for age in x]
plt.scatter (x,y, c=color)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.show()
```

```python
x = titanic['Age']
y = titanic['Fare']

# Drop NaN values to avoid mismatches
x = x.dropna()
y = y.dropna()

# Align x and y indices to ensure they match
x, y = x.align(y, join='inner')

# Create a color list based on age
color = [
    'red' if age > 65 else 
    'green' if age > 55 else 
    'blue' if age > 45 else 
    'black' if age > 35 else 
    'brown' if age > 25 else 
    'pink' 
    for age in x
]

# Scatter plot with color argument corrected
plt.scatter(x, y, c=color)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.show()
```

```python
class_counts = titanic['Pclass'].value_counts()

class_counts.plot(kind='bar', color=['blue', 'green', 'black'])

plt.xlabel('Passenger class')
plt.ylabel('Number of passengers')
plt.title('Number of passenger by class')

plt.show()
```

```python
class_counts = titanic['Survived'].value_counts()

class_counts.plot(kind='bar', color=['red', 'green'])

plt.xlabel('Survivers')
plt.ylabel('Number of passengers')
plt.title('Survived vs Non Survived')
plt.xticks(ticks=(0, 1), labels=['Survived', 'Non Survived'])
plt.show()
```
