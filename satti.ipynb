```python
df=pd.read_csv("titanic.csv")
```

```python
x=df['Age']
y=df['Fare']
```

```python
plt.scatter(x,y, edgecolors='red', facecolor='green')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
```

```python
x = df['Age']
y = df['Fare']
color = ['red' if age > 30 else 'green' for age in x]
plt.scatter (x,y, c=color)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
plt.show()
```
