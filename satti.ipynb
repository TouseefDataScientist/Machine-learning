```python
class_counts= df['Pclass'].value_counts()
class_counts.plot(kind='bar', color=['red', 'green', 'blue'] )
plt.xlabel('Passenger class')
plt.ylabel('Number of passengers')
plt.title('Servived')
plt.show()
```

```python
#create a bar chart
class_counts= df['Survived'].value_counts()
class_counts.plot(kind='bar', color=['red', 'green',] )
plt.xlabel('surviver')
plt.ylabel('Number of passengers')
plt.title('Servived vs non survised')
plt.xticks(ticks=(0, 1), labels=['survised', 'Non survised'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.hist (df['Age'].dropna(), bins=20,color='red', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
```

```python
plt.figure(figsize=(7,8))
plt.subplot(1,2,1)
plt.hist (df['Fare'].dropna(), bins=20,color='red', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
```
