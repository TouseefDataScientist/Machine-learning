```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic_data = sns.load_dataset('titanic')
```

```python
titanic_data.head()
```

```python
titanic_data.info()
```

```python
plt.figure(figsize=(10, 7)) 
sns.histplot(titanic_data['age'].dropna(), bins=30, kde=True, color='red')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.barplot(x='sex', y='survived', data=titanic_data)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='class', data=titanic_data)
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='class', y='survived', data=titanic_data)
plt.title('Survival Rate by Class')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='age', y='survived', hue='sex', data=titanic_data)
plt.title('Survival Rate by Age and Gender')
plt.xlabel('Age')
plt.ylabel('Survival')
plt.show()
```
