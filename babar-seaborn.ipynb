```python
plt.figure(figsize=(10,6))
sns.violinplot(x='day', y='total_bill', hue='time' ,data=tips, )

plt.title('Violin Plot of Total Bill by Day and Time')
plt.xlabel('Day')
plt.ylabel('Total Bill')   
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='day', y='total_bill', hue='sex' ,data=tips, )

plt.title('Violin Plot of Total Bill by Day and Time')
plt.xlabel('Day')
plt.ylabel('Total Bill')   
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='day', y='total_bill', hue='sex' ,data=tips, )

plt.title('Bar Plot of Sex')
plt.xlabel('Day of the Week')
plt.ylabel('count of diners')   
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='day',  hue='smoker' ,data=tips )

plt.title('Count of Diners by Day and Smoking Preference')
plt.xlabel('Day of the Week')
plt.ylabel('Count of Diner')   
```
