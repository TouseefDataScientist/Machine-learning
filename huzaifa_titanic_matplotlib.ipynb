```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
df.isnull().sum()
```

# Drop Columns

```python
df = df.drop(['Name','Ticket','Cabin'],axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

# Fill null values

```python
#Fill null values in Age column
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Age'].isnull().sum()
```

```python
#Fill null values in Embarked
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df['Embarked'].isnull().sum()
```

```python
df.isnull().sum()
```

# Scatter plot of Age and Fare

```python
x = df['Age']
y = df['Fare']

plt.scatter(x,y,marker='*')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
x = df['Age']
y = df['Fare']

plt.scatter(x,y, c='red', label='Age')
plt.scatter(x,y, c='green', label='Fare', alpha=0.5)

plt.xlabel('Age')
plt.ylabel('Fare')

plt.title('This is Age and Fare Scatter')
plt.legend(['Age (red)', 'Fare(green)'])
plt.legend(loc='upper left')
plt.show()
```

# Histogram of Age and Fare

```python
plt.figure(figsize=(10,5))

#plotting histogram for Age
plt.subplot(1,2,1)
plt.hist(df['Age'].dropna(),bins=20,color='red',edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')

#plotting histogram for Fare
plt.subplot(1,2,2)
plt.hist(df['Fare'].dropna(),bins=20,color='green',edgecolor='black')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Histogram of Fare')

plt.plot()
```

# Survived vs Not survived

```python
status = df['Survived'].value_counts()

plt.xlabel('class')
plt.ylabel('Frequency')
plt.title('Analysis of Survived and Notsurvived')
status.plot(kind='bar',color = ['red','green'])
plt.xticks(ticks=[0,1], labels=['Not Survived','Survived'],rotation=45)
plt.show()
```

# Age vs Survived

```python
survived = df[df['Survived'] == 1]
not_survived = df[df['Survived'] == 0]
plt.figure(figsize=(14,14))

plt.subplot(2,3,1)
plt.hist(survived['Age'],bins=20,label='Survived',color='green')
plt.title('Age Distribution of Survived')
plt.xlabel('Age')
plt.ylabel('Passenger')

plt.subplot(2,3,2)
plt.hist(not_survived['Age'],bins=20,alpha=0.5,label='Not Survived',color='red')
plt.title('Age Distribution of Not Survived')
plt.xlabel('Age')
plt.ylabel('Passenger')
plt.show()
```

# Survived vs Gender

```python
gender_survived = df.groupby(['Sex', 'Survived']).size()
gender_survived.plot(kind='bar',figsize=(10,6),color=['red','green'],alpha=0.5)
plt.title("Survival by Gender")
plt.xlabel('Gender')
plt.ylabel("Number of passangers")
xtick_labels = [f'({sex}, {"Not Survived" if survived == 0 else "Survived"})' for sex, survived in gender_survived.index]
plt.xticks(ticks=range(len(xtick_labels)), labels=xtick_labels, rotation=45, ha='right', fontsize=10)
#plt.xticks(ticks=[0,1], labels=['Not Survived','Survived'],rotation=0)
plt.legend(['Not Survived', 'Survived'], title="Survived")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# PClass

```python
p_class = df['Pclass'].value_counts()

plt.xlabel('P_Class')
plt.ylabel('Frequency')
plt.title('Bar graph of P class')
p_class.plot(kind='bar',color = ['blue','green','red'])
plt.xticks(rotation=0)
plt.show()
```

# Pie Charts

# PClass

```python
class_count = df['Pclass'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(class_count,labels=class_count.index,autopct='%1.1f%%',colors=['green','skyblue','purple'])
plt.legend()
plt.show()
```

# Survived vs Not Survived

```python
class_count = df['Survived'].value_counts()
plt.figure(figsize=(8,8))
mylabel = ['Not Survived',"Survived"]
plt.pie(class_count,labels=mylabel,autopct='%1.1f%%',colors=['red','green'])
plt.legend()
plt.show()
```

# Gender Visualization

```python
gender_pie_chart = df['Sex'].value_counts()
plt.figure(figsize=(8,8))

gender_label = ['Female','Male']
plt.pie(gender_pie_chart,labels=gender_label,autopct='%1.1f%%',colors=['seagreen','brown'],explode=[0.01,0.01])
plt.legend()
plt.show()
```

# Sibling and Spouse Visualization

```python
sib_sp = df['SibSp'].value_counts()

plt.xlabel('Sibling and Spouse')
plt.ylabel('Frequency')
plt.title('Bar graph of Sibling and Spouse')
sib_sp.plot(kind='bar',color = ['blue','green'])
plt.xticks(rotation=0)
plt.show()
```

```python
sib_sp = df['SibSp'].value_counts()

plt.figure(figsize=(8,8))
myexplode = [0.01,0.01,0.1,0.1,0.1,0.1,0.01]
plt.pie(sib_sp,labels=sib_sp.index,autopct='%1.1f%%',colors=['red','green','blue','skyblue','gray','orange','purple'],explode=myexplode)
plt.legend()
plt.show()
```

# Embarked
